```python
from matplotlib.ticker import MultipleLocator
import matplotlib.pyplot as plt
import pandas as pd

files = ['DIN.csv', 'DIEN.csv', 'BST.csv','LightSANs.csv', 'KG-DSI_n_plus.csv', 'KG-DSI_n_avg.csv']
names = ['DIN', 'DIEN', 'BST','LightSANs', 'KG-DSI(sum)', 'KG-DSI(mean)',]
hists = {}
colors = ['red', 'blue', 'green','purple', 'orange', 'black']

i = 0
for file in files:
    hists[names[i]] = pd.read_csv(
        './ml/' + file,
        index_col=0
    )
    i += 1

for i in hists:
    print(i)
    print('%.4f'%hists[i]['test_auc'].max())

for i in hists:
    print(i)
    print('%.4f'%hists[i]['test_acc'].max())


x_major_locator=MultipleLocator(1)

# AUC
plt.figure(figsize=(8, 6), dpi=200)
ax=plt.gca()
ax.xaxis.set_major_locator(x_major_locator)
plt.title('AUC comparison on MovieLens-1M Dataset')
j = 0
for i in hists:
    if names[j] in ['LightSANs', 'KG-DSI(sum)', 'KG-DSI(mean)']:
        plt.plot(list(range(20)),hists[i]['test_auc'][1::2],color=colors[j])
        for a, b in zip(hists[i][:20].index, hists[i]['test_auc'][1::2]):
            plt.text(a, b, '%.4f' % b, ha='center', va='bottom', fontsize=6, color=colors[j])
    else:
        plt.plot(list(range(20)),hists[i]['test_auc'][:20],color=colors[j])
        for a, b in zip(hists[i][:20].index, hists[i]['test_auc'][:20]):
            plt.text(a, b, '%.4f' % b, ha='center', va='bottom', fontsize=6, color=colors[j])

    j += 1

plt.legend(names)
plt.grid(linestyle=":", color="black", zorder=0)  # 
plt.show()

# ACC
plt.figure(figsize=(8, 6), dpi=200)
ax=plt.gca()
ax.xaxis.set_major_locator(x_major_locator)
plt.title('Acc comparison on MovieLens-1M Dataset')
j = 0
for i in hists:
    plt.plot(list(range(20)),hists[i]['test_acc'][:20],color=colors[j])
    for a, b in zip(hists[i][:20].index, hists[i]['test_acc'][:20]):
        plt.text(a, b, '%.4f' % b, ha='center', va='bottom', fontsize=6, color=colors[j])
    j += 1

plt.legend(names)
plt.grid(linestyle=":", color="black", zorder=0)  # 
plt.show()

# plt.figure(figsize=(8, 5), dpi=200)
#
# plt.plot(me['test_acc'])
# plt.title('v1 test acc')
# for a, b in zip(me.index, me['test_acc']):
#     plt.text(a, b, '%.4f' % b, ha='center', va='bottom', fontsize=10, color='red')

# plt.show()
```
